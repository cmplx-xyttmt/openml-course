```python
def is_married(status):
    married_string = 'Married'
    len_married = len(married_string)
    if len(status) < len_married:
        return False
    return status[0:len_married] == married_string

df_married_men = adult_data[(adult_data['sex'] == 'Male') & (adult_data['marital-status'].apply(lambda status: is_married(status)))]
df_unmarried_men = adult_data[(adult_data['sex'] == 'Male') & (adult_data['marital-status'].apply(lambda status: not is_married(status)))]
```

```python
df_married_men['salary'].value_counts(normalize=True)
```

```python
df_unmarried_men['salary'].value_counts(normalize=True)
```

#### Some Notes
There are different types of features:
- Numeric (e.g age, salary). Expressed as floats
- Categoric (e.g City, Service Type) Expressed as strings.
- Ordinal features. Limited number of possible values and there is an order. (e.g education (primary, secondary, university)). Expressed as string (strings can be mapped to ints)
- Binary features. (e.g 0/1, male/female) can represented as strings and floats.
